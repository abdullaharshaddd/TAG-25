```python
# Step 1: Fetch a News Article
import requests
import google.generativeai as genai

# Your API keys here
GUARDIAN_API_KEY = "eafc3f40-20c4-42f8-99f8-ec8c5dd2c1bf"
GEMINI_API_KEY = "AIzaSyBLgGq0fAR75XWTi4WJeZ584qbOKqh0ALk"

# Configure Gemini
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel("gemini-2.5-flash-preview-04-17")

# Fetch article from The Guardian
guardian_url = "https://content.guardianapis.com/search"
params = {
    "q": "technology",   # Change topic if you want
    "order-by": "newest",
    "api-key": GUARDIAN_API_KEY,
    "page-size": 1,
    "show-fields": "bodyText,headline"
}

response = requests.get(guardian_url, params=params)
data = response.json()

# Extract the first article's content
article = data["response"]["results"][0]
headline = article["webTitle"]
article_text = article["fields"]["bodyText"]

# Step 2: Gemini Prompt
prompt = f"""
You are a professional news anchor. Your job is to rewrite the following article into a natural-sounding script that can be read aloud on a news broadcast.

Please follow these rules:
- Use short, clear, spoken-style sentences.
- Keep a friendly, neutral, and engaging tone.
- Start with a strong opening line to grab attention.
- Avoid long paragraphs or overly technical words.
- Simplify the language while keeping facts accurate.
- End with a natural closing like ‚ÄúThis is AI News, bringing you the latest.‚Äù

Here is the article content:
{article_text}
"""

# Generate script with Gemini
response = model.generate_content(prompt)
news_script = response.text

# Output
print("üì∞ Headline:", headline)
print("\nüéôÔ∏è Anchor Script:\n")
print(news_script)
```
